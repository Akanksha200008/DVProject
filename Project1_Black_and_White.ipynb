```python
import pandas as pd
data= pd.read_csv("Police Shootings Age Race.csv")
data.head(10)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Filtered data for black people
black_people_data= data[data['race'] == 'B']

# Create a larger figure size
plt.figure(figsize=(10, 6))

# Plot a smooth approximation of the histogram with kernel density estimate (KDE)
sns.histplot(data=black_people_data['age'], kde=True, color='RED', edgecolor='black')
plt.xlabel('Age',fontsize=12)
plt.ylabel('Frequency',fontsize=12)
plt.title('Age Distribution for Black People with KDE')
# Add a grid for better readability
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Use tight layout to avoid cutting off parts of the graph
plt.tight_layout()

plt.show()
```
